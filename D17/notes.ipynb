/*
-- Create Department table -----------------------
CREATE TABLE Department (
    id INT PRIMARY KEY,
    name VARCHAR(50)
);

-- Create Employees table ------------------------
CREATE TABLE Employees (
    id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    city VARCHAR(50),
    hiring_year INT,
    FOREIGN KEY (department_id) REFERENCES Department(id)
);
*/

-- View Department table --------------------------
SELECT * FROM Department;

-- Insert data into Department --------------------
INSERT INTO Department (id, name)
VALUES
    (1, 'Art'),
    (2, 'Religion'),
    (3, 'Science');

-- View Employees table ---------------------------
SELECT * FROM Employees;

-- Insert data into Employees ---------------------
INSERT INTO Employees (id, name, department_id, salary, city, hiring_year)
VALUES
    (1, 'Aafiyah', 1, 2500, 'Sepang', 2021),
    (2, 'Rania',   2, 2000, 'KL',     2009),
    (3, 'Aydan',   3, 4000, 'KL',     2017);

-- Query employees by city -------------------------
SELECT name, salary, city, hiring_year
FROM Employees
WHERE city IN ('New York', 'Chicago');

-- To filter result table --------------------------
SELECT name, salary, city, hiring_year
FROM Employees
WHERE salary BETWEEN 2000 AND 5000
ORDER BY hiring_year DESC;

-- To add a column table but the column is not saved -----------

SELECT name, salary, salary * 12 AS yearly_salary
FROM employees;


-- To sort the order --------------------------------
SELECT * FROM Employees

SELECT name, salary
FROM employees
ORDER BY salary ASC;
