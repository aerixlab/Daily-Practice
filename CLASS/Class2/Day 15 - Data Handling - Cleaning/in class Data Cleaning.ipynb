```python
# newdf['Price'].fillna(0 , inplace=True)

# we are using a central tendency to fill the nan values

newdf = df.copy()

mean_price = newdf['Price'].median()
print(mean_price)


newdf['Price'].fillna(mean_price, inplace=True)
newdf
```

#### outliers in your data, instead of using mean, you can use median, can be a better representative of the central tendency

```python
print(newdf['Price'].sort_values())
```

```python
# column fillna() # mean median mode[0] 


squared = newdf['Quantity'] ** 2 # square everything

newQuantity = squared / (squared**0.5) # divide the squares with their roots

newQuantity

newdf['Quantity'] = newQuantity
newdf

median_qnty = newdf['Quantity'].median()

newdf['Quantity'].fillna(median_qnty, inplace=True)
newdf
```
