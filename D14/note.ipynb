```python
import numpy as np

A = np.array(
[[2,3,4],
[6,7,8],
[3,2,3]]
)

B = np.array(
[[3,3,3],
[3,3,3],
[3,3,3]]
)
```

```python
multiply = B @ A
print(multiply)
# Cannot print A @ B coz different dimension
```

```python
# @ multiply like matrices while, * times each value
print('\nMatrix Multiplication: \n',A @ B)
print('\nSimple Array Multiplication: \n',A * B)

print('\nMatrix Multiplication: \n',B @ A)
print('\nSimple Array Multiplication: \n',B * A)
```
