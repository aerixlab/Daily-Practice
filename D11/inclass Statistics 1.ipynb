```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# how many unique values do you see in all the columns, and what can you infer
# 28 value \\ 19 unique
# what is the average bill amount: 
# what is the average tip amount
# just find some new insight: for every rm20 spent, they got tip rm3
```

```python
# 1st way
average_bill = sum(df.total_bill) / len(df.total_bill)
print(average_bill)

average_tip = sum(df.tip) / len(df.tip)
print(average_tip)
```

```python
# 2nd way
average = df['total_bill'].sum() / df['total_bill'].__len__()
print(average)

tip = df['tip'].sum() / df['tip'].__len__()
print(tip)
```
